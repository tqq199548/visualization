```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy

from PIL import Image
import requests
from io import BytesIO
```

```python
x=np.array(["Luffy","Zoro","Nami","Usopp","Sanji"])
y2=np.array([170,100,90,120,50])

response=requests.get("https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png")
image_file=BytesIO(response.content)
image=Image.open(image_file)
```

```python
fig,ax=plt.subplots(figsize=(7,3))
ax.plot(x,y2,color="red")
```

```python
fig,ax=plt.subplots(figsize=(7,3))
ax.plot(x,y2,color="red")
####^^^above are basic figure
ax_image=fig.add_axes([0.1,0.11,0.15,0.35])    ###figure the position with additional axis
```

```python
import pandas as pd

# Sample data based on the archived Podio page
data = {
    'Name': ['Honoré de Balzac', 'John Milton', 'Haruki Murakami', 'Wolfgang Amadeus Mozart', 'Ludwig van Beethoven'],
    'Sleep': [5, 6, 6, 5, 8],
    'Creative Work': [8, 6, 5, 12, 8],
    'Day Job/Admin': [0, 0, 0, 0, 0],
    'Food/Leisure': [3, 4, 4, 3, 2],
    'Exercise': [1, 2, 4, 1, 1],
    'Other': [7, 6, 5, 3, 5]
}

df = pd.DataFrame(data)

import matplotlib.pyplot as plt

# Set the 'Name' column as the index
df.set_index('Name', inplace=True)

# Plotting the stacked bar chart
ax = df.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='tab20')
plt.title('Daily Routines of Famous Creative Individuals')
plt.xlabel('Individuals')
plt.ylabel('Hours Spent')
plt.legend(loc='upper left', bbox_to_anchor=(1,1), title='Activities')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import  PIL
import requests
```

```python
tips=sns.load_dataset("tips")
print(tips)
```

```python
sns.lineplot(data=tips,
             x='total_bill',
             y='tip')
```

```python
sns.set_style('whitegrid')

tipgraph=sns/lineplot(data=tips,
                      x='total_bill',
                      y='tip')

tipgraph.set(title = 'Tips vs Total Bill',
             xlabel = 'Total Bill ($)',
             ylabel = 'Tip Amount ($)')
```

```python
fig=plt.subplots(figsize=(10,3))

tipgrah=sns.lineplot(data=tips,
                     x='total_bill',
                     y='tip',
                     color='hotpink',
                     linestyle='--',
                     linewidth=3,
                     marker='o',
                     markerfacecolor='indigo')
tipgrah.set(title='Tips vs Total Bill',
            xlabel='TOTAL bILL($)',
            ylabel='TIP Amount($)')
```

```python
tipgraph=sns.scatterplot(data=tips,
                         x='total_bill',
                         y='tip',
                         style='time',
                         hue='day',
                         palette=['purple','hotpink','deepskyblue','yellowgreen'])

tipgrah.set(title='Tips VS Total Bill',
            xlabel='Total Bill($)',
            ylabel='Tip Amount ($)')
```

```python
sns.pairplot(data=tips,hue='day')
```

```python
daysplot=sns.relplot(data=tips,
                    x='total_bill',
                    y='tip',
                    hue='sex',
                    col='day',
                    kind='scatter',
                    col_wrap=2)
daysplot
```

```python
import plotly.graph_objects as go 

x1=np.array(['Luffy','Zoro','Nami','Usopp','Sanji'])
y1=np.array([110,180,240,99,220])
```

```python
graph=go.Figure()
graph.add_trace(go.Bar(x=x1,y=y1))
graph.update_layout(title='Pirate Scores',
                    xaxis_title="Pirtes",
                    yaxis_title="Socre")
```

```python
graph=go.Figure()

graph.add_trace(go.Scatter(x=x1,
                           y=y1,
                           mode='markers',
                           marker=dict(
                               size=15,
                               color='hotpink',
                               opacity=1,
                               line=dict(width=5,color='purple')
                           )))

graph.update_layout(title='Interactive Pirate Plot',
                    xaxis_title='Pirates',
                    yaxis_title='Scores',
                    width=500,
                    height=500)
```

```python
from wordcloud import WordCloud

df=pd.read_csv("https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv",
               on_bad_lines='skip')
df
```

```python
text=" ".join(each for each in df.quote)

wordcloud=WordCloud(background_color='White',
                    colormap='inferno').generate(text)

fig,ax=plt.subplots(figsize=(7,3))
ax.imshow(wordcloud)

ax.axis("off")
```

```python
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted

A=set(["apple","banana","watermelon"])
B=set(["pumpkin","blueberry","apple","Key lime"])

diagram=venn2([A,B],
              set_labels=('Fruits','Pires'),
              set_colors=('blue','red'),
              alpha=0.5)

diagram.get_label_by_id("10").set_text("/n".join(A-B))    #LEFT
diagram.get_label_by_id("11").set_text("/n".join(A&B))   #MIDDLE
diagram.get_label_by_id("01").set_text("/n".join(B-A))     #RIGHT
```
